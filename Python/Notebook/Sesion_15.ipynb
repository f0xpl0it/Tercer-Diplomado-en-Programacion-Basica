For small data sets you might be able to replace the wrong data one by one, but not for big data sets.

To replace wrong data for larger data sets you can create some rules, e.g. set some boundaries for legal values, and replace any values that are outside of the boundaries.

#### Example

Loop through all values in the "Duration" column.

If the value is higher than 120, set it to 120:

````python
import pandas as pd

df = pd.read_csv('data.csv')

for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120

print(df.to_string())
````

##  **Removing Rows**

This way you do not have to find out what to replace them with, and there is a good chance you do not need them to do your analyses.

Example

````python
import pandas as pd

df = pd.read_csv('data.csv')

for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)

#remember to include the 'inplace = True' argument to make the changes in the original DataFrame object instead of returning a copy

print(df.to_string())
````

<div class="alert alert-block" style="background-color: #FFD08A; color: black;">
<b>Note:</b> The <code>corr()</code> method ignores "not numeric" columns.

###  *Result Explained*

The Result of the `corr()` method is a table with a lot of numbers that represents how well the relationship is between two columns.

The number varies from -1 to 1.

1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.

0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.

-0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.

<div class="alert alert-block" style="background-color: #FFD08A; color: black;">
<b>What is a good correlation?</b> It depends on the use, but I think it is safe to say you have to have at least <code>0.6</code> (or <code>-0.6</code>) to call it a good correlation.

###  *Perfect Correlation:*

We can see that "Duration" and "Duration" got the number `1.000000`, which makes sense, each column always has a perfect relationship with itself.

###  *Good Correlation:*

"Duration" and "Calories" got a `0.922721` correlation, which is a very good correlation, and we can predict that the longer you work out, the more calories you burn, and the other way around: if you burned a lot of calories, you probably had a long work out.

###  *Bad Correlation:*

"Duration" and "Maxpulse" got a `0.009403` correlation, which is a very bad correlation, meaning that we can not predict the max pulse by just looking at the duration of the work out, and vice versa.

# **Data Filtering**
---




In many datasets, extreme values can affect our analysis.  
We will use the **Interquartile Range (IQR)** method to remove values that are too far from most of the data.

The `filter_outliers()` function allows filtering a specific column using a factor that controls how strict we are in removing extreme values.

#### Example


````python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("data.csv")  # replace with your file path

# Show first rows
df.head()
````

##  **Function to Filter Outliers**

This function filters out extreme values in a column using the IQR method.


#### Example

````python
def filter_outliers(df, col, factor=1.5):
    """
    Removes outliers from a dataframe column using the interquartile range (IQR).

    Parameters:
    df : pd.DataFrame
        DataFrame to filter
    col : str
        Column name
    factor : float
        Multiplier for the IQR to determine limits

    Returns:
    pd.DataFrame
        Filtered DataFrame
    """
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - factor * IQR
    upper_limit = Q3 + factor * IQR

    filtered_df = df[(df[col] >= lower_limit) & (df[col] <= upper_limit)]
    return filtered_df
````

##  **Applying the Filter to All Numeric Columns**




We will apply the filter to all numeric columns in the dataset using a factor of 1.5.

#### Example

````python
numeric_cols = df.select_dtypes(include='number').columns

df_filtered = df.copy()
for col in numeric_cols:
    df_filtered = filter_outliers(df_filtered, col, factor=1.5)

# Compare the number of rows
print("Original rows:", len(df))
print("Rows after filtering:", len(df_filtered))
````

##  **Density Plot (Kernel Density Estimate)**

A density plot shows the **distribution of a numeric column** as a smooth curve.  
It helps us understand where the data is concentrated and visualize the effect of filtering outliers.

#### Example

````python
for col in numeric_cols:
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    # Before filtering
    axes[0].boxplot(df[col].dropna())
    axes[0].set_title(f"{col} - Before Filtering")

    # After filtering
    axes[1].boxplot(df_filtered[col].dropna())
    axes[1].set_title(f"{col} - After Filtering")

    plt.tight_layout()
    plt.show()
````

##  **Visualize Distributions With Seaborn**



Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.



#### Example

````python
import matplotlib.pyplot as plt
import seaborn as sns

cols_to_plot = numeric_cols.tolist()
n_rows = 2
n_cols = 2

plt.figure(figsize=(18, 10))  # tamao total de la figura

for i, col in enumerate(cols_to_plot):
    plt.subplot(n_rows, n_cols, i+1)
    sns.kdeplot(df[col].dropna(), label='Before', fill=True)
    sns.kdeplot(df_filtered[col].dropna(), label='After', fill=True)
    plt.title(f"{col} - Density Comparison")
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.legend()

plt.tight_layout()
plt.show()
````
